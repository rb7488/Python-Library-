**NumPy**

```
#revision
import numpy as np
a=np.array([1,2,3,4,5])
print(a)
print(type(a))
```

```
b=np.array([[1,2,3],[4,5,6]])
print(b)
print("Shape: ",b.shape) #shape tells the row and column available
```

```
#indexing
print(b[0,0])
print(b[1,2])
print(b[:,1])
print(b[0])

```

```
#basic operationn
arr=np.array([10,20,30,40])
print(arr+5)
print(arr*2)
print(arr/10)
print(np.sum(arr))
print(np.mean(arr))
```

```
#reshape
x=np.array([1,2,3,4,5,6])
x_reshaped=x.reshape(2,3)
print(x_reshaped)
print("New Shape: ",x_reshaped)
```

```
#slice
arr=np.array([10,20,30,40,50,60])

print(arr[1:4])
print(arr[:3])
print(arr[-3:])
```

```
#practise question
ar=np.array([5,10,15,20,25,30])
print(ar)
print(ar*3)
print(ar+5)
new_ar=ar.reshape(2,3)
print(new_ar)
print("New array: ",new_ar)
print(new_ar[1])
print(new_ar[:2])
```

Data Visualization in Matplot

```
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[1,4,9,16,25,36]
#creating line plot
plt.plot(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')
plt.show()
```

```
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[1,4,9,16,25,36]
#creating line plot
plt.plot(x,y,color='red',linestyle='--',marker='o') #marker used for coorinate for x
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')
plt.show()
```

```
#multiple plots
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y1=[1,4,9,16,25,36]
y2=[2,4,6,8,10,12]
plt.figure(figsize=(9,5))
plt.subplot(2,2,1) #used for ek page pr bahut se visual represent 2 part me divide
plt.plot(x,y1,color='red')
plt.show()
plt.subplot(2,2,2)
plt.plot(x,y2,color='green')
plt.show()
plt.subplot(2,2,3)
plt.plot(y1,y2,color='blue')
plt.show()
```
