```python

import multiprocessing
import time
def sq_number():
    for i in range(5):
     time.sleep(1)
     print(f"Square: {i*i}")
def cube_number():
    for i in range(5):
     time.sleep(1.5)
     print(f"Cube :{i*i*i}")
if __name__=="__main__":

 #create two process (p1 = ek naya process jo sq_number() run karega,

# p2 = ek alag process jo cube_number() run karega,dono diff process,diff memory,OS scheule this)


    p1=multiprocessing.Process(target=sq_number)
    p2=multiprocessing.Process(target=cube_number)
    t=time.time()
    #start the process(dono kaam saath chalao, parallel execution start hota hai.)
    p1.start()
    p2.start()
    #waiting for the process to complete(jab tk p1,p2 complete n ho jye)
    p1.join()
    p2.join()
```

```python
#multiple threading with threadpool executor
from concurrent.futures import ThreadPoolExecutor
import time
def print_num(num):
    time.sleep(1)
    return f"number:{num}"
num=[1,2,3,4,5,6]
with ThreadPoolExecutor(max_workers=3) as executor:
    results=executor.map(print_num,num)
    for result in results:
        print(result)
```

```python
#multiple process with processpool executor
from concurrent.futures import ProcessPoolExecutor
import time
def sq_num(num):
    time.sleep(1)
    return f"number:{num*num}"
num=[1,2,3,4,5,6]
with ProcessPoolExecutor(max_workers=3) as executor:
    results=executor.map(sq_num,num)
    for result in results:
        print(result)
```
