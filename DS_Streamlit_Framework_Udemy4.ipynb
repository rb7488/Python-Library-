'''Streamlit is an open-source Python framework used to build interactive web applications for data science and machine learning.
 It allows developers to create data dashboards, visualizations, and ML model demos directly from Python code,
 without requiring knowledge of HTML, CSS, or JavaScript.'''




```python
#app.py basic
import streamlit as st
import pandas as pd
import numpy as np
#title of appication
st.title("Hello Streamlit")
#Display a simple text
st.write("This is a simple text")
#create a simple dataframe
df=pd.DataFrame({
 'first column':[1,2,3,4],
 'second column':[10,20,30,40]
})
#display the data frame
st.write("Here is the data frame")
st.write(df)
#create a line chart
chart_data=pd.DataFrame(
    np.random.randn(20,3),columns=['a','b','c']
)
st.line_chart(chart_data)
```

```python
#widegt.py on vs code
import streamlit as st
import pandas as pd
import numpy as np

st.title("Streamlit Text Input")
name=st.text_input("Enter your name: ")
age=st.slider("Select your age", 0,100,20)
st.write(f"Your age is: {age}")

options=["Python","Java","C++","Javascript"]
choice=st.selectbox("Choose your fav language: ",options)
st.write(f"You selected {choice}.")
if name:
    st.write(f"Hello, {name} welcome to your page")

data={
    "Name":["John","Jake","Bhaskar","Krish"],
    "Age":[28,34,21,34],
    "City":["New Yorl","London","New Delhi","Noida"]
}
df=pd.DataFrame(data)
df.to_csv("sampledata.csv")
st.write(df)

uploaded_file=st.file_uploader("Choose a CSV file",type="csv")
if uploaded_file is not None:
    df=pd.read_csv(uploaded_file)
    st.write(df)


```
