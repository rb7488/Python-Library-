````python
from flask import Flask
''' it create an instance of Flask class,
which will be our WSGI (web server gateway interface) application.
'''
app = Flask(__name__)
@app.route('/')
def hello_world():
  return "Hello Welcome to flask"

@app.route('/index')
def index():
  return "Hello! welcome on index page"

if __name__=="__main__":
    app.run(debug=True)
````

````python
#Integrating HTML with Flask
app = Flask(__name__)
@app.route('/')
def hello_world():
  return "<html><h1>Hi Welcome to Bhaskar Baithaki</h1></html> "

if __name__=="__main__":
    app.run(debug=True)
#html code is large so we write to other page and render template
````

````python
from flask import Flask,render_template
app = Flask(__name__)
@app.route('/')
def hello_world():
  return "<html><h1>Hi Welcome to Bhaskar Baithaki</h1></html> "
@app.route('/index')
def index():
  return render_template('index.html') # index.html is your html file
if __name__=="__main__":
    app.run(debug=True)
````

````python
#get and post
from flask import Flask, request

app = Flask(__name__)

@app.route("/hello", methods=["GET"])
def hello():
    name = request.args.get("name")
    return f"Hello {name}"
#post
from flask import Flask, request

app = Flask(__name__)

@app.route("/submit", methods=["POST"])
def submit():
    data = request.form["name"]
    return f"Received: {data}"
    #running the main
if __name__=="__main__":
    app.run()
````
