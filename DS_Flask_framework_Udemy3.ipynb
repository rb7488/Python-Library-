````python
from flask import Flask
''' it create an instance of Flask class,
which will be our WSGI (web server gateway interface) application.
'''
app = Flask(__name__)
@app.route('/')
def hello_world():
  return "Hello Welcome to flask"

@app.route('/index')
def index():
  return "Hello! welcome on index page"

if __name__=="__main__":
    app.run(debug=True)
````
